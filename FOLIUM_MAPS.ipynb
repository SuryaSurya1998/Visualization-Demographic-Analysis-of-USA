```
df_ind = df.set_index('Place_Code')
df_ind
```

```
df_neigh = df_ind[['CENTLAT',	'CENTLON', 'State_Name', 'County_Name']]
df_neigh
```

```
df_neigh = df_neigh[~df_neigh.index.duplicated(keep='first')]
df_neigh
```

```
df1=store.loc[:,['StoreID','Latitude','Longitude']]
df1
```

```
np.shape(df1)
```

```
tree = BallTree(np.deg2rad(df1[['Latitude', 'Longitude']].values),leaf_size =20, metric='haversine')
```

```
query_lats = df_neigh['CENTLAT']
query_lons = df_neigh['CENTLON']


# use k = 5 for 5 closest neighbors

distances, indices = tree.query(np.deg2rad(np.c_[query_lats, query_lons]), k = 20)
```

```
store_list=[]
lat_list =[]
long_list =[]
for store, i, lat, log in zip(df1['StoreID'], indices, df1['Latitude'], df1['Longitude']):
  store_list.append(df1['StoreID'][i].values)
  lat_list.append(df1['Latitude'][i].values)
  long_list.append(df1['Longitude'][i].values)
```

```
s1=pd.Series(store_list,name='StoreId')
s2=pd.Series(lat_list,name='Latitude')
s3=pd.Series(long_list ,name='longitude')
prox_df = pd.concat([s1,s2,s3], axis=1)

prox_df
```

```
prox_df_plo =prox_df.apply(pd.Series.explode).reset_index()
prox_df_plo
```

```
#MARKING ALABAMA IN MAP

#map = folium.Map(location=[32.3182, -86.9023], tiles="Stamen Toner", zoom_start=10)
map = folium.Map(location=[32.3182, -86.9023], min_zoom= 5)

# adding the current location
folium.CircleMarker(location=[32.3182, -86.9023],radius=5,color='red',fill=True,fill_color='red').add_to(map)
folium.Marker(location=[32.3182, -86.9023],popup="<strong> Current Location</strong>",icon = folium.Icon(color='red')).add_to(map)

#drawing circle with radius provided to search
folium.Circle(location=[32.3182, -86.9023],radius=15000,color='blue',opacity=0.5,fill=True,fill_color='blue').add_to(map)
map.save("MARKING ALABAMA IN MAP.html")
map
```
